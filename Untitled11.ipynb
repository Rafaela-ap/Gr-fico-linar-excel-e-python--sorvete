```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("sakshisatre/ice-cream-sales-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import glob
import numpy as np

# Assuming 'path' contains the directory where the dataset files are located
# Find the CSV file in the directory
csv_files = glob.glob(os.path.join(path, '*.csv'))

# Check if any CSV files were found
if not csv_files:
    raise FileNotFoundError("No CSV files found in the dataset directory.")

# Assuming there's only one CSV file or we should take the first one
df = pd.read_csv(csv_files[0])

display(df.head())
```

```python

# Extracting the data from the DataFrame
x = df['Temperature']
y = df['Revenue']

# Calculate the linear regression line
coef = np.polyfit(x, y, 1)
line = np.poly1d(coef)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data Points', alpha=0.6)
plt.plot(x, line(x), color='red', label=f'Regression Line: y = {coef[0]:.2f}x + {coef[1]:.2f}')

plt.title('Linear Regression: Ice Cream Sales vs. Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.legend()
plt.show()

print(f"Equação da linha: y = {coef[0]:.2f}x + {coef[1]:.2f}")
```
